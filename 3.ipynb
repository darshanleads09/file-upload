```python
# Practical 3: Multiclass Classification (FFNN)

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf

# 1. Load dataset
data = load_iris()
X = data.data        # features
y = data.target      # target (0,1,2)

# 2. Convert target to categorical (IMPORTANT)
y = tf.keras.utils.to_categorical(y)

# 3. Split data
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25)

# 4. Build model
model = Sequential()
model.add(Dense(8, input_dim=4, activation='relu'))   # hidden layer 1
model.add(Dense(6, activation='relu'))                # hidden layer 2
model.add(Dense(3, activation='softmax'))             # output layer

# 5. Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 6. Train model
model.fit(xtrain, ytrain, epochs=50, batch_size=32)

# 7. Predict
pred = model.predict(xtest)

# 8. Show first 5 predictions
for i in range(5):
    print("Predicted:", pred[i], "Actual:", ytest[i])
```
